```python
import math

n = 1
print(f"Последовательность a_n = 2**n-n")
a = 2**n-n
print(f"1й элемент = {a}")
for i in range (2,10):
    a = 2**(n*i)-(n*i)
    print(f"{i}й элемент = {a}")

    

n = 2
print(f"Последовательность b_n = 1/(1-n)")
b = 1/(1-n)
print(f"1й элемент = {b}")
for i in range (2,10):
    b = 1/(1-n*i)
    print(f"{i}й элемент = {b}")

n = 1
print(f"Последовательность c_n = -1**n+sqrt(2n)")
c = -1**n+math.sqrt(2*n)
print(f"1й элемент = {b}")
for i in range (2,10):
    c = -1**(n*i)+math.sqrt(2*(n*i))
    print(f"{i}й элемент = {c}")

n = 1
print(f"Последовательность d_n = (-1)**(2*n)+1/n**2)")
d = (-1)**(2*n)+ 1/(n**2)
print(f"1й элемент = {d}")
for i in range (2,10):
    d = (-1)**(2*(n*i))+ 1/((n*i)**2)
    print(f"{i}й элемент = {d}")
```
